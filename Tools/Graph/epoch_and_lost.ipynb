```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
model_final = pd.read_csv('Data/model_final.csv')
model_final_norm = pd.read_csv('Data/model_final_norm.csv')
model_final_covid_learn = pd.read_csv('Data/covid_learn.csv')
```

```python
print(model_final.head())
print(model_final_norm.head())
print(model_final_covid_learn.head())
```

```python
model_final['loss'].plot(kind='line', legend=True, color='#C2185B')
model_final['val_loss'].plot(kind='line', legend=True, color='#42a5f5')
plt.xlim((-1, 50))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.savefig('Saves/final_loss.eps', format='eps', dpi=1200)
```

```python
model_final['val_acc'].plot(kind='line', color='#00897b')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.xlim((-1, 50))
plt.ylim((0.2, 1))
plt.scatter(49, 0.9358, s=5, label='Test Accuracy')
plt.legend()
#plt.annotate('0.9358', xy=(49, 0.9358)).set_rotation(-90)
plt.savefig('Saves/final_acc.eps', format='eps', dpi=1200)
```

```python
model_final['val_acc'].plot(kind='line', color='#00897b')
model_final_norm['val_acc'].plot(kind='line', color='#fb8c00', label="val_acc_normalised")
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.xlim((-1, 50))
plt.ylim((0, 1))
plt.legend()
#plt.annotate('0.9358', xy=(49, 0.9358)).set_rotation(-90)
plt.savefig('Saves/final_acc_norm.eps', format='eps', dpi=1200)
```

```python
model_final_covid_learn.set_index('epoch')
model_final_covid_learn.plot(kind='line', x='epoch', y='val_acc', color='#00897b')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.xlim((48, 70))
plt.ylim((0.85, 0.96))
plt.scatter(49, 0.9358, s=25, label='Test Accuracy #1')
plt.scatter(69, 0.9422, s=25, label='Test Accuracy #2')
plt.xticks(ticks=[49, 59, 69])
plt.legend()
#plt.annotate('0.9358', xy=(49, 0.9358)).set_rotation(-90)
plt.savefig('Saves/covid_learn_acc.eps', format='eps', dpi=1200)
```
